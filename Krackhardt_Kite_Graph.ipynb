# Krackhardt Kite Graph with NetworkX
This notebook creates and visualizes the Krackhardt Kite graph using NetworkX and Matplotlib.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()

nodes = ["Andre", "Beverly", "Carol", "Diane", "Ed", "Fernando", "Garth", "Heather", "Ike", "Jane"]
G.add_nodes_from(nodes)

edges = [
    ("Andre", "Beverly"), ("Andre", "Carol"), ("Andre", "Diane"),
    ("Beverly", "Diane"), ("Beverly", "Ed"),
    ("Carol", "Fernando"), ("Carol", "Diane"),
    ("Diane", "Fernando"), ("Diane", "Garth"),
    ("Fernando", "Garth"), ("Fernando", "Heather"),
    ("Garth", "Ed"),
    ("Heather", "Ike"), ("Ike", "Jane")
]
G.add_edges_from(edges)
```

```python
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G, seed=42)
nx.draw(
    G, pos, with_labels=True, node_color='lightblue', edge_color='purple',
    node_size=2000, font_size=10, font_weight='bold'
)
plt.title("Krackhardt Kite Graph", fontsize=16)
plt.show()
```
